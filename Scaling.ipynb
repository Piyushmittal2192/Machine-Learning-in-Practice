```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(1, 100, 50)
x.std()
```

```python
min_max_norm = (x - x.min())/(x.max() - x.min())
z_score_norm = (x - x.mean())/x.std()
print("Standard Deviation after Min-Max Normalization: {}".format(min_max_norm.std()))
print("Standard Deviation after z-score Normalization: {}".format(z_score_norm.std()))
```

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 2.5))
axs[0].scatter(np.linspace(1, 50),x)
axs[0].set_title('Unscaled')
axs[1].scatter(np.linspace(1, 50),min_max_norm)
axs[1].set_title('Min-Max Normalization')
axs[2].scatter(np.linspace(1, 50),z_score_norm)
axs[2].set_title('z-score Normalization')
```
